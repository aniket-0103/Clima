# Initialising the parameters

```python
def initialise(dim):
    w=np.random.randn(dim,1)*0.01      # w and b are  the inputs
    b=np.zeros((1,1))
    #print(w.shape)
    #print(b.shape)
    return w,b
```

```python
def sigmoid(z):      # Function to calculate activation function 
    a=1/(1+np.exp(-z))
    return a
```

```python
print(Y.shape)
print(X.shape)
```

# Calculate the predicted value of y

```python
def y_pred(X,w,b):
    w,b=initialise(X.shape[1])
    z=(np.dot(X,w))+b
    a=sigmoid(z)
    #print(a)
    return a
```

# Calculate cost function

```python
def cost_func(X,w,b):
    a=y_pred(X,w,b)
    l=Y*np.log(a)+(1-Y)*np.log(1-a)
    l=(-1/m)*np.sum(l,axis=0)
    #print(l)
    return l
```

```python
w,b=initialise(X.shape[1])
a=y_pred(X,w,b)
print(a.shape)
print(X.shape)
print(Y.shape)
```

# Code the gradient descent 

```python
def gradient_descent(a,Y,X,w,b,lr):
    dw0=sum(np.dot((a-Y).T,X[:,0]))
    dw1=sum(np.dot((a-Y),X[:,1]))
    db=(a-Y)*1
    j=0
    while(True):
        print(w)
        w0=w[0]-dw0*lr
        w1=w[1]-dw1*lr
        b0=b-lr*db
        w_new=[w0,w1]
        print(w_new)
        print(cost_func(X,w,b))
        if (j>500):
            return w_new,b0
        j+=1
    
```

```python
w,b=initialise(X.shape[1])
a=y_pred(X,w,b)
w,b0=gradient_descent(a,Y,X,w,b,0.01)
```

```python
def graph(formula):
    x=x1
    y=formula()
    plt.plot(x,y)
    
def my_formula():
    return (-b0-(w[0]/100)*x1)/w[1]/100

fig, ax=plt.subplots()
sns.scatterplot(x1,x2,hue=Y)
graph(my_formula)
plt.xlabel('x_0')
plt.ylabel('x_1')
plt.show()
    
```
